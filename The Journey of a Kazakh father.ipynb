```python
# importing random to be used in functions.
import random 

##############################################################################
# Some repeting choices in script.
# \N is used for new line when """ is not used, then new line in triple " is used.
# To allow for better flow of text, <Press enter/return to continue/try again> with input function
# is used repetedly
# Prompt decleared in input functions as directed, and give instructions for input,
# or in most cases to direct the users attention to the input area using ">"
##############################################################################

#introduction ahead of the game.
print("""The Kazakhstan government
(the following should be read with a Borat accent):

Daughter Owner’s Manual

Official publication of Ministry of Agriculture and Wildlife

Our condolences, we see that your wife has given you a daughter.
Fortunately for you, your diminished nation’s government has developed a tool
for electrical calculator that can guide you and your daughter. """)

# Input function to start game.
input(prompt = "<Press enter/return to continue>\n")

# Defining a stage of the game as a function.
# Stage 1.
def room_1():
    """
    A) This game is inspired by the movie Borat 2. And takes 
    inspiration from Borat 1.
    Quotes and text are taken from the Borat universe and do not represent 
    my opinions.
    It consist of five stages, where the first two is a “tool” to help fathers. 
    The following three stages is a story line of a father giving away her daughter.
    It has defined functions to help facilitate the different outcomes of answers.
    This game requires some knowledge of the Borat movies.But it also has some 
    input of random chance to make the game replayable.

    Stage 1: What your daughter needs to grow up.
    Stage 2: Answer questions about what women can do.
    Stage 3: Decide your daughters future.
    Stage 4: Fulfill your daughters destiny.
    Stage 5: Win your daughter back.

    B) Known Bugs and/or Errors:
    Potentially repeating same questions in stage 2.
    Exit can lead to repeat of while loop.
    """
    
    print("""Finding the right way to accommodate your daughter is never easy. 
But they need clear boundaries.
""")
    
    input(prompt = "<Press enter/return to continue>")
    
    print( 
"""
Which of these would make the best home for your daughter growing up?
    1. A dog cage.
    2. A horse cage.
    3. A bedroom.
    4. I can't afford any of them.
        
""")
        
    confinement = input(prompt = ">")
    
    # Converting object to lowercase with casefold to facilitate conditional statement.
    confinement = confinement.casefold()

    # developing a conditional statement.
    if confinement == "1" or "dog" in confinement:
        print("A dog cage is not enough space. Try again. ")
        room_1() # This moves us back to repeat question.

    elif confinement == "2" or "horse" in confinement:
        print("""Great job, it is important that your daughter has the necessary 
amount of space to grow up the right way within the necessary confinement. """)
        input(prompt = "<Press enter/return to continue>\n")
        room_2() # This moves us into next question.


    elif confinement == "3" or "bedroom" in confinement:
        print("""You are on a dangerous path. The diminished nation of 
Kazakhstan cannot accept this to be standard practice for women. """)
        replay() # This moves to fail, and asks if you want to replay.
    
    elif confinement == "4" or "afford" in confinement:
        print("""Your government understands, another acceptable option could 
be to tie your daughter in a shed. """)
        input(prompt = "<Press enter/return to continue>\n")
        room_2() # This moves us into next question.
        
    else:
        print("Try again, and please use one of the options. ")
        input(prompt = "<Press enter/return to continue>\n")
        room_1() # This moves us back to repeat question.
        
# Questions are defined as functions to allow repeat of question from this point.        

# Defining stage 2.
def room_2():
    print("""Your daughter has grown older and is starting to ask questions. 
Now you need to answer some test question to show that you can answer your daughter. """)
    input(prompt = "<Press enter/return to continue>\n")
    room_2_q1()
    
# Establishing function to allow for return to start of questions.    
def room_2_q1():
    
    #Question list, including only last part of question. "" indicate direct movie quotes
    question_list = [ 
"own businesses? ", 
"drive cars? ", 
"be journalists? ",
"touch themselves? ",
"be loved like sons? ", 
"decide who they marry? ",
'"be educate?" ', ]
    
    # Establishing number of questions to be asked. And setting value to subtract from.
    questions = 2
    
    # While loop to keep player in questioning until two right answers in a row.
    while questions > 0:
        question = random.choice(question_list)
        print("Question:")
        print("Can women " + question) # "Can women" is set as start to any 
        # question from list, and rest of question added from question_list.
        print("")

        answer = input(prompt = "Answer yes or no. >")
        
        # Using casefold 
        answer = answer.casefold()

        if answer == "no" or "n" in answer:
            print("Correct, of course.\n ")
            # When correct answer is given one is subtracted from number of questions.
            questions -=1

        elif answer == "yes" or "y" in answer:
            print("""This is very illegal and can create great harm to our nation.\n """)
            room_2_q1() # return to question start and restart of number of questions.

        else:
            # In case of unknown input.
            print("""Error, please answer "yes" or "no". """)

    print("""Congratulations! You have completed the training set up by your 
diminished nation.
""")
    
    input(prompt = "<Press enter/return to continue>\n")
    room_3() # Moving to new stage.

# New stage after guidance, you have now moved to life with your daughter.
def room_3():    
    print("""It is time to get your daughter ready for marriage, she is 
approaching 15 years old and will soon be the oldest unmarried woman in all 
Kazakhstan.
""")
    print("""Where do you want to send your daughter?

    1. To brother.
    2. To local boy.
    3. To new fancy golden cage in the US and A.
    4. Set her free. 
""")
    
    destination = input(prompt = ">")
    destination = destination.casefold() # Converting using casefold
    
    if destination == "1" or "brother" in destination:
        print("""Your son becomes modestly happy. She is no longer your problem.

You have spent your life raising this girl, looking at her playing in her 
confinement since she was little. Throwing her your scraps and silencing here 
every time she said something. This was the wrong decision, unfortunately your 
wife gave you another daughter. You can try again.
""")
        input(prompt = "<Press enter/return to continue>\n")
        room_3()
        
        
    elif destination == "2" or "local" in destination or "boy" in destination:
        print("""The local boy initially becomes very happy. But she only gave
him daughters, and following Kazakh tradition he ended the relationship by ending her. 
""")
              
        print("""You have spent your life raising this girl, looking at her 
playing in her confinement since she was little. Throwing her 
your scraps and silencing here every time she said something. 
This was the wrong decision, unfortunately your wife gave you 
another daughter. You can try again.""")
              
        input(prompt = "<Press enter/return to continue>\n")
        room_3()
    
    elif destination == "3" or "golden" in destination or "us" in destination or "usa" in destination:
        print("""
You are in America. You want to give your daughter a better life!
""")
    
        print("""To get her ready for her golden cage like Melania Trump, you first 
need to take her to man who can help.
""")
        print("""Going to the plastic surgeon office, you need to decide how many 
dollars you want to spend, up to 22000, enter amount in digits.
""")
        america()
        
        
    elif destination == "4" or "free" in destination:
        print("""Congratulations, your daughter is following in your sisters 
footsteps, she is now number four prostitute in all Kazakhstan, pushing her 
aunt down to fifth. Your daughter is a self-made modern woman, a true Kazakh 
feminist. You are so proud! """)
              
        print("""Now let us see how well you can do with your next daughter, 
because there can only be one number four prostitute in Kazakhstan. And you
don't want to remove your other daughter from that position. """)
              
        input(prompt = "<Press enter/return to continue>\n")
        room_3()



def america():
    dollars = input(prompt = ">")
    
    # Controlling for wrong type of input.
    try:
        # Converting the object to integer.
        dollars = int(dollars)
    
    # Request for new input if last input did not meet requirements.
    except:
        print('Not what I was looking for, please input number without "," or ".".')
        input(prompt = "<Press enter/return to continue>\n")
        america()    
       
    # wonton: dialog taken from movie scene
    if dollars == 21751:
        print("""
Coming out of the clinic, your daughter is invited to a grand ball.              
There Fat King Donald sees your daughter,
he becomes more turgid than he’d ever been before...

...So he grabs her "vagine"...

Donald and your daughter then waltz; he still has her in the vag-hold.

And they live happily ever after. The end!
""")
        print("""You found the ultimate victory for you daughter, 
but you still have other daughters. """)
        input(prompt = "<Press enter/return to continue>\n")
        room_3()
    
    
    elif dollars == 0:
        print("""So you don't want to help your daughter get the best life 
possible, she will not get any golden cage in America, and will be oldest 
unmarried woman in all of Kazakhstan. """)
        print("""You have spent your life raising this girl, looking at her 
playing in her confinement since she was little. Throwing her 
your scraps and silencing here every time she said something. 
This was the wrong decision, unfortunately your wife gave you another 
daughter. You can try again.""")
        
        input(prompt = "<Press enter/return to continue>\n")
        room_3()
    
    elif dollars >= 10000:
        
        # Start of nested conditional. To deal with potential higher number.
        if dollars > 22000:
            print("""I am keeping everything over 22,000 dollars. 
Thank you for the extra tip. """)
            plastic = 3
            
        else:
            plastic = 3
        
    elif dollars >= 5000:
        plastic = 2
        
    elif dollars < 5000:
        plastic = 1

    else:
        print("Error, please input number without , or .") # Failsafe in case except fail.
        america()
        
        
    print("You are ready to take your daughter to the clinic. ")
    input(prompt = "<Press enter/return to continue>\n")
    print("""Your daughter has learned things in America, she no longer wishes 
to be given away to a man. She now believes women can accomplish things on their own. """)

    
    print( 
"""
Are you sure you want to go through with this? (yes or no)
        
""")
    
    surgery = input(prompt = ">")
        
    surgery = surgery.casefold()
    # developing a conditional statement
    if surgery == "yes" or "y" in surgery:
        print("The surgery was a success, now let us see who she can attract. ")
        input(prompt = "<Press enter/return to continue>\n")
        husband = random.randint(1, 4)
        
        # Start of nested conditional to keep variable in function.
        # Random husband results, with likelyhood of husband option impacted 
        # by the amount of money spent in previous question.
        if husband > plastic:
            print("""Lucky you, your daughter has managed to seduce Michael Pence. 
"The Vice Premier was known to be such a pussy-hound that he could not be left 
alone in a room with a woman" but your daughter will now be. 
""")
            
            print("But was this what you wanted for her? ")
            winback() # Sending to next stage.
            
        elif husband == plastic:
            print("""To find a man your daughter knows where she needs to go. 
Rudy Giuliani is an avid landscaper, and Four Seasons Total Landscaping is his 
favorite place to hang out. She immediately finds him there and seduces him. 
""")
            print("But was this what you wanted for her? ")
            winback()
            
        elif husband < plastic:
            print("""This is okay, she manages to seduce Jeffrey Epstein. 
She was initially too old for him, but luckily Ghislaine Maxwell helps your 
daughter to get close to him. 
""")
            print("But was this what you wanted for her? ")
            winback()
            
        else:
            print("Error") # Failsafe
           

    elif surgery == "no" or "n" in surgery:
        print("""You let her find her own way, not hearing from her after you
call off the surgery...
        
She still decides to go for a man in the US & A. But without
"help" her only option is Jeffrey Epstein. Luckily, she is still young enough 
at 15 to be one of the oldest girls at Epstein's Island. 
""")
        print("But was this what you wanted for her? ")
        
        input(prompt = "<Press enter/return to continue>\n")
        winback()
    
    # Deals with non accepted input. Because of need to keep variabel,
    # the user is told to input number from previous question.
    else:
        print("""Error, please answer "yes" or "no". But I am afraid I forgot
How much you wanted to spend. Please input number. """)
        america()
            

def winback():
    print("You still have a chance to win her back. ")
    
    print( 
"""
What do you want to do?
    1. What do you mean? She is other man's problem now.
    2. Trade places with her and let her live her life.
    3. Ask new man to release her.    
""")
    

    back = input(prompt = ">")

    back = back.casefold()
    
    
    if back == "1" or "?" in back or "problem" in back:
        print("Have you learned nothing... ")
        replay()

    # Option that wins the game and follows the story from Borat, dialog is
    # taken directly from Borat.
    elif back == "2" or "trade" in back: 
        print("""You find out where she is seducing her American man, and go 
there to make the trade:
You burst into the room wearing Lingerie.
    YOU: She too old for you. She 15!
    MAN: What, why are you dressed like this?
    YOU: She’s my daughter. Take me instead. Take my anoos.
    MAN: I don’t want you.
    YOUR DAUGHTER: No! Don’t take him.
    YOU: No, take my anoos. Do not have her.
    YOU: I better, my back pussy very tight.
    YOUR DAUGHTER: No, please, my front anoos.
Your daughter and you start fighting in Kazakh. As her American man starts to 
walk out...
    MAN: Oh my God, what are you doing?
    YOU: Listen, I was in prison many years. I have techniques with my mouth.
    MAN: What’s goin’ on here? Look at this guy.
    YOU: I forbid this union. No, Trump will be disappoint. You are 
    leaving hotel without golden shower.
American man leaves hotel never to be heard from again. 

    YOUR DAUGHTER: You know what I think? I think you did this because you love 
    me as much as your sons.
    YOU: No.
""")
        input("<Press enter/return to continue>")
        print("""
    YOU: More. So much more.
A smile crosses your daughters face. This is all she ever
wanted to hear.

""")
        input(prompt = "<Press enter/return to continue>\n")
        print("""Your newfound love for your daughter helps lead a change in 
Kazakhstan, together you change the whole nation.""") 
        input(prompt = "<Press enter/return to continue>\n")
        print("""Kazakhstan now feminist nation! Like US & A and Saudi Arabia. 
Bride exports declared misogynist so we now traffic grooms. Kevin Spacy 
is now Kazakhstan’s biggest trade partner. """)
        input(prompt = "<Press enter/return to continue>\n")
        print("""Congratulation, you have grown and become a great man. 
""")
        win_replay()
            
    elif back == "3" or "ask" in back or "release" in back:
        print("""When you ask the American man to return your daughter, 
he does not even know who she is. You have lost her, you might have believed 
that she ended up with him, but she ended up on Epstein's Island like all the 
other girls anyway... """)
        replay()
    
    else:
        print("Error, please try again. ")
        winback()

# Defined function allowing you to replay once you have failed.
def replay():
    
    print("""You've failed, thank you for playing! Do you want to play again? 
(yes or no) """)
    
    answer = input(prompt = ">")
    
    answer = answer.casefold()
    
    if answer == "yes" or "y" in answer:
        print("")
        input(prompt = "<Press enter/return to continue>\n")
        room_1()

    elif answer == "no" or "n" in answer:
        input(prompt = "<Press enter/return to exit>\n")
        
    else:
        print("""Error, please answer "yes" or "no". """)
        replay()
# Replay function when game is won
def win_replay():

    print("""Thank you for playing! Do you want to play again? 
(yes or no) """)
    answer = input(prompt = ">")
    answer = answer.casefold()
    
    if answer == "yes" or "y" in answer:
        print("")
        input(prompt = "<Press enter/return to continue>\n")
        room_1()

    elif answer == "no" or "n" in answer:
        input(prompt = "<Press enter/return to exit>\n")
        
    else:
        print("""Error, please answer "yes" or "no". """)
        winreplay()

# calling the function to start game.
room_1()
```
